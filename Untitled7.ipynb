```python
from qiskit import QuantumCircuit, execute, BasicAer
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt

# Number of CSPs and Users
num_csp = 2
num_users = 2
total_qubits = num_csp + num_users

# Create a quantum circuit with qubits for CSPs and users
qc = QuantumCircuit(total_qubits, total_qubits)

# Initialize qubits to represent the initial state |ψ⟩
for i in range(total_qubits):
    qc.h(i)  # Apply Hadamard gate to create a superposition state

# Define the strategy unitary operations for CSPs and Users
theta_csp = [np.pi/4, np.pi/4]  # Strategy angles for CSPs
theta_user = [np.pi/3, np.pi/3]  # Strategy angles for Users

# Apply rotation gates for CSPs and Users
for i in range(num_csp):
    qc.ry(theta_csp[i], i)  # Apply Ry rotation for CSPs

for j in range(num_users):
    qc.ry(theta_user[j], num_csp + j)  # Apply Ry rotation for Users

# Add measurements to all qubits
qc.measure_all()

# Print the circuit structure
print("Circuit Structure with Strategy Rotations:")
print(qc)

# Use BasicAer for simulation
backend = BasicAer.get_backend('qasm_simulator')

# Execute the circuit
num_shots = 1000
job = execute(qc, backend, shots=num_shots)
result = job.result()

# Get and display the counts
counts = result.get_counts(qc)
print("\nMeasurement Counts:")
for state, count in sorted(counts.items()):
    print(f"|{state}⟩: {count} times")

# Create matplotlib visualization
plt.figure(figsize=(12, 6))

# Extract states and counts for plotting
states = list(counts.keys())
values = list(counts.values())

# Create color map based on CSP and User states
colors = []
for state in states:
    csp_part = state[:num_csp]
    user_part = state[num_csp:]
    if csp_part.count('1') > csp_part.count('0'):
        colors.append('lightblue')
    else:
        colors.append('lightgreen')

# Create bar plot
bars = plt.bar(range(len(states)), values, color=colors, alpha=0.7)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# Customize the plot
plt.title('Quantum State Measurements with Strategy Rotations', fontsize=14, pad=20)
plt.xlabel('Quantum States (CSPs | Users)', fontsize=12)
plt.ylabel('Counts', fontsize=12)

# Set x-axis ticks with state labels
plt.xticks(range(len(states)), states, rotation=45)

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Add a legend explaining the colors
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='lightblue', alpha=0.7, label='CSP Majority 1s'),
    Patch(facecolor='lightgreen', alpha=0.7, label='CSP Majority 0s')
]
plt.legend(handles=legend_elements)

plt.tight_layout()
plt.show()

# Create Qiskit-style histogram
plt.figure(figsize=(12, 6))
plot_histogram(counts,
              color=['lightblue', 'lightgreen', 'pink', 'orange', 'purple', 'yellow'],
              title='Quantum Strategy State Distribution')
plt.tight_layout()
plt.show()

# Calculate and display statistics
total_counts = sum(counts.values())
print("\nStrategy Statistics:")
print(f"Total measurements: {total_counts}")
print(f"Number of unique states: {len(counts)}")
print(f"Expected number of states: {2**total_qubits}")
print("\nStrategy Angles:")
print("CSP Rotations:", [f"{theta:.3f} radians ({theta*180/np.pi:.1f}°)" for theta in theta_csp])
print("User Rotations:", [f"{theta:.3f} radians ({theta*180/np.pi:.1f}°)" for theta in theta_user])
print("\nState Probabilities:")
for state, count in sorted(counts.items()):
    prob = count/total_counts
    # Split state into CSP and User parts
    csp_part = state[:num_csp]
    user_part = state[num_csp:]
    print(f"|{state}⟩ (CSPs: {csp_part}, Users: {user_part}): {prob:.3f}")

# Create additional visualization for strategy angles
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.pie([1], labels=['CSP Angles'], colors=['lightblue'], startangle=90)
plt.title('CSP Strategy Angles')
for i, theta in enumerate(theta_csp):
    plt.text(0, 0-i*0.5, f'CSP {i+1}: {theta*180/np.pi:.1f}°',
             ha='center', va='center')

plt.subplot(122)
plt.pie([1], labels=['User Angles'], colors=['lightgreen'], startangle=90)
plt.title('User Strategy Angles')
for i, theta in enumerate(theta_user):
    plt.text(0, 0-i*0.5, f'User {i+1}: {theta*180/np.pi:.1f}°',
             ha='center', va='center')

plt.tight_layout()
plt.show()
```

```python
from qiskit import QuantumCircuit, execute, BasicAer
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt

def update_strategy(theta, delta):
    """Update the strategy by adjusting the angle."""
    return theta + delta

# Number of CSPs and Users
num_csp = 2
num_users = 2
total_qubits = num_csp + num_users

# Optimization loop parameters
delta = np.pi/20  # Step size for updating strategy
max_iterations = 10

# Define the strategy unitary operations for CSPs and Users
theta_csp = [np.pi/4, np.pi/4]  # Initial strategy angles for CSPs
theta_user = [np.pi/3, np.pi/3]  # Initial strategy angles for Users

# Use BasicAer for simulation
backend = BasicAer.get_backend('qasm_simulator')

# Store results for analysis
all_counts = []
all_angles = []

# Optimization loop
for iteration in range(max_iterations):
    # Reset the circuit for each iteration
    qc = QuantumCircuit(total_qubits, total_qubits)

    # Initialize qubits
    for i in range(total_qubits):
        qc.h(i)

    # Introduce dynamic entanglement between CSPs and Users
    for i in range(num_csp):
        for j in range(num_users):
            qc.cx(i, num_csp + j)  # Apply CNOT gate for entanglement

    # Apply updated strategies
    for i in range(num_csp):
        theta_csp[i] = update_strategy(theta_csp[i], delta)
        qc.ry(theta_csp[i], i)

    for j in range(num_users):
        theta_user[j] = update_strategy(theta_user[j], delta)
        qc.ry(theta_user[j], num_csp + j)

    # Measure the qubits
    qc.measure_all()

    # Execute the circuit
    result = execute(qc, backend, shots=2048).result()
    counts = result.get_counts()

    all_counts.append(counts)
    all_angles.append((theta_csp.copy(), theta_user.copy()))

# Get final results
final_counts = all_counts[-1]

# Print the final strategy angles
print("\nFinal Strategy Angles:")
print("CSP Rotations:", [f"{theta:.3f} radians ({theta*180/np.pi:.1f}°)" for theta in theta_csp])
print("User Rotations:", [f"{theta:.3f} radians ({theta*180/np.pi:.1f}°)" for theta in theta_user])

# Create figure with subplots
fig = plt.figure(figsize=(15, 6))

# Plot 1: Custom Histogram
ax1 = plt.subplot(1, 2, 1)
states = list(final_counts.keys())
probabilities = [count/2048 for count in final_counts.values()]
bars = ax1.bar(range(len(states)), probabilities)
ax1.set_xticks(range(len(states)))
ax1.set_xticklabels(states, rotation=45)
ax1.set_title('Final Quantum Measurement Results')
ax1.set_xlabel('Quantum State')
ax1.set_ylabel('Probability')

# Add probability values on top of bars
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}',
             ha='center', va='bottom')

# Plot 2: Strategy Evolution
ax2 = plt.subplot(1, 2, 2)
iterations = range(max_iterations)
csp_angles = [np.mean([angles[0][0], angles[0][1]]) for angles in all_angles]
user_angles = [np.mean([angles[1][0], angles[1][1]]) for angles in all_angles]
ax2.plot(iterations, csp_angles, 'b-', label='CSP Strategy')
ax2.plot(iterations, user_angles, 'r-', label='User Strategy')
ax2.set_xlabel('Iteration')
ax2.set_ylabel('Average Angle (radians)')
ax2.set_title('Strategy Evolution')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# Print measurement statistics
print("\nFinal Measurement Statistics:")
total_shots = sum(final_counts.values())
for state, count in sorted(final_counts.items()):
    probability = count/total_shots
    # Split state into CSP and User parts
    csp_part = state[:num_csp]
    user_part = state[num_csp:]
    print(f"State |{state}⟩ (CSPs: {csp_part}, Users: {user_part}): {count} shots ({probability:.3%})")
```
